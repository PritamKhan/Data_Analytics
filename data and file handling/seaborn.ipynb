```python
sns.relplot(x="total_bill", y="tip", data=df, hue= 'smoker' )
#hue give the different color shede to the data points based on the smoker column
#insite >> customer who smoke gave the highest tip
```

```python
sns.relplot(x="total_bill", y="tip", data=df, hue= 'size')
```

```python
sns.relplot(x="total_bill", y="tip", data=df, style= 'size')
#style give the different shape/representation to the data points based on the size column
#insite >> customer who are in group of 3 gave the highest tip
```

```python
sns.relplot(x="total_bill", y="tip", data=df, style= 'smoker', hue= 'time')
```

```python
#plot between day and total_bill, day is categorical data
sns.catplot(x = 'day', y = 'total_bill', data = tips, )
#insite >> on weekend the bill amount is more as compared to other days 
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
#swamr plot is used to show the distribution of categorical data by using dots without overlapping
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip)
#joint plot is used to show the relationship between two variables and their distribution
#insite >> more the bill amount more the tip
#and also disctribution of both the variables
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = 'reg')
```
