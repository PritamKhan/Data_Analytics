```python
#merge
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
df1
```

```python
def creat_flag(x):
    if x < 10:
        return 'cheap'
    if x >= 10 and x < 20:
        return 'medium'
    else:
        return 'high'
```

```python
df["Flag_Fare"] = df['Fare'].apply(creat_flag)
```
